```python
# Given that a passenger survived, what are the odds that they are male?

# Given: All Survivors
# Event: Male Passengers

survivors_df = titanic_df[titanic_df['Survived'] == 1 ]
num_survivors = len(survivors_df)
num_survivors

prob_survived = num_survivors /len(titanic_df)
prob_survived

males_df = survivors_df[survivors_df['Sex'] == "male"]
prob_male_survived_intersect = len(males_df) / len(titanic_df)
prob_male_survived_intersect

prob_male_survived_intersect / prob_survived

survivors_df.describe()
```
