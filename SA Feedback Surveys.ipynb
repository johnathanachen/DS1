```python
COLS_TO_RENAME = {
    "How would you rate your overall satisfaction with the Summer Academy this week?": "Satisfaction",
    "How well is the schedule paced?": "Schedule Pace",
    "How well are the tutorials paced?": "Tutorial Pace", 
    "What track are you in?": "Track",
    "How likely is it that you would recommend the Make School Summer Academy to a friend?": "Recommend"
}

def places():
    places = []
    for survey in configfiles:
        place = re.search(r'-(.*?)\.csv', survey).group(1)
        places.append(place)
    return places


def create_dataframes():
    frames = []
    for survey in configfiles:
        title = re.search(r'-(.*?)\.csv', survey).group(1)
        name = title+"_df"
        name = pd.read_csv(survey)
        name.columns.name = title
        frames.append(name)
    return frames

def rename_fields(frames):
    edited_frames = [frame.rename(index=str, columns=COLS_TO_RENAME) for frame in frames] 
    return edited_frames



    
places = places()
frames = create_dataframes()
edited_frames = rename_fields(frames)
edited_frames
```
